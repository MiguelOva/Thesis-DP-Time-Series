```python
def plot_history(history):
    # plot history
    pyplot.plot(history.history['loss'], label='train')
    pyplot.plot(history.history['val_loss'], label='test')
    pyplot.legend()
    pyplot.title('Mean Squared Error between Train and Test Sets')
    pyplot.show()
```

```python
def prophet_function(data,periods):
    # Renaming the column names accroding to Prophet's requirements
    prophet_df = data.reset_index()[['Date','Adj_Close']]
    prophet_df.rename(columns={'Date':'ds','Adj_Close':'y'},inplace=True)
    prophet_df['ds'] = prophet_df['ds'].dt.tz_localize(None)

    # Basic model 
    prophet_basic = Prophet()
    prophet_basic.fit(prophet_df[['ds','y']])
    
    # Making predictions for 1 year, until 2022-01-01
    future= prophet_basic.make_future_dataframe(periods=periods)
    forecast_df = prophet_basic.predict(future)
    columns = ['ds', 'trend', 'yhat_lower', 'yhat_upper', 'weekly', 'weekly_lower','weekly_upper', 'yhat']
    display(forecast_df[columns].tail())
    
    # Plot predictions
    fig_1 = prophet_basic.plot(forecast_df)
    a = add_changepoints_to_plot(fig_1.gca(), prophet_basic, forecast_df)
    print('Change points: ','{prophet_basic.changepoints}\n')
```
